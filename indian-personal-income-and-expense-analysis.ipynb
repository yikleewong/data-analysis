```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = '/kaggle/input/indian-personal-finance-and-spending-habits/data.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.groupby(['Occupation','Age']).mean(numeric_only =True)
```

```python
ax1 = df.pivot_table(values='Income', index='Age', columns='Occupation', aggfunc='mean').plot(kind='line', figsize=(10,6)) 
ax1.set_title('Income by Age and Occupation') 
ax1.set_xlabel('Age') 
ax1.set_ylabel('Income') 
ax1.grid(True) 
ax1.legend(loc = "best",bbox_to_anchor = (1, 1))
```

```python

plt.figure(figsize=(15, 12))

plt.subplot(221)
sns.histplot(df.groupby('Occupation')['Income'].get_group('Student'), bins=30, kde=True, color='red')
plt.title('Student Income Distribution')
plt.subplot(222)
sns.histplot(df.groupby('Occupation')['Income'].get_group('Retired'), bins=30, kde=True, color='orange')
plt.title('Retired Income Distribution')
plt.subplot(223)
sns.histplot(df.groupby('Occupation')['Income'].get_group('Self_Employed'), bins=30, kde=True, color='green')
plt.title('Self_Employed Income Distribution')
plt.subplot(224)
sns.histplot(df.groupby('Occupation')['Income'].get_group('Professional'), bins=30, kde=True, color='blue')
plt.title('Professional Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(x = df['Age'],y = df['Rent'],alpha=0.5,c='blue',hue=df['City_Tier'])
plt.show()
```

```python
ax2 = df.pivot_table(index = "Dependents", columns = "City_Tier", values = "Rent", aggfunc = "mean").plot.bar(rot=0)
df.groupby('Dependents')['Rent'].mean().plot(label='mean rent', c='black', marker='o', ax=ax2)
plt.legend(loc = "best", bbox_to_anchor = (1, 1))
plt.show()
```

```python
expenses = ['Rent', 'Loan_Repayment', 'Insurance','Groceries', 'Transport', 'Eating_Out', 'Entertainment',
'Utilities', 'Healthcare', 'Education', 'Miscellaneous']
num_expenses = len(expenses)

expenses_palette = sns.color_palette("Set2", num_expenses)

fig, axes = plt.subplots(4, 3, figsize=(20, 25))
axes = axes.flatten()

for ax, expense, color in zip(axes, expenses, expenses_palette):
    sns.histplot(df[expense], bins=50, kde=True, color=color, ax=ax)
    ax.set_title(f'{expense} Distribution', fontsize=16)
    ax.set_xlabel(f'{expense} (INR)', fontsize=14)
    ax.set_ylabel('Number of Users', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
pie_data = df.groupby('City_Tier').mean(numeric_only =True)
fig, axes = plt.subplots(1, 3, figsize=(20, 8)) 
fig.suptitle('Expense Distribution by City Tier', fontsize=16) 
for i, tier in enumerate(pie_data.index):
    axes[i].pie(pie_data.loc[tier, expenses], labels=expenses, autopct='%1.1f%%', startangle=140) 
    axes[i].set_title(f'{tier}') 
plt.show()
```
